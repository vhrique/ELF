# Packages

```python
import os
import torch

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from glob import glob
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
from torch.utils.data import Dataset
```

# Data loading

```python
data_path = '/content/drive/MyDrive/Estudos/Datasets/CMAPSSData'
test_files = list(glob(os.path.join(data_path, 'test_FD00*.txt')))
train_files = list(glob(os.path.join(data_path, 'train_FD00*.txt')))
```

```python
experiment_columns = ['unit', 'cycle']
condition_columns = [f'cond_{i+1}' for i in range(3)]
feature_columns = [f'sensor_{i+1}' for i in range(21)]
column_names = experiment_columns + condition_columns + feature_columns
dfs = []
for filename in train_files:
    df = pd.read_csv(filename, sep=' ', names=column_names, index_col=False)
    df['group'] = os.path.split(filename)[-1]
    df['RUL'] = None
    df['cycle_max'] = None
    dfs.append(df)
train_df = pd.concat(dfs).reset_index()
```

```python
dfs = []
for filename in test_files:
    df = pd.read_csv(filename, sep=' ', names=column_names, index_col=False)
    df['group'] = os.path.split(filename)[-1]
    df['RUL'] = None
    df['cycle_max'] = None
    dfs.append(df)
test_df = pd.concat(dfs).reset_index()
```

## Adjusting target (RUL)

```python
for group in train_df.group.unique():
    df_group = train_df.query(f'group == "{group}"')
    for unit in df_group.unit.unique():
        df_unit = df_group.query(f'unit == {unit}')
        mask = (train_df.group == group) & (train_df.unit == unit)
        train_df.loc[mask, 'cycle_max'] = df_unit.cycle.max()
train_df.loc[:, 'RUL'] = train_df.cycle_max - train_df.cycle
```

```python
for group in test_df.group.unique():
    df_group = test_df.query(f'group == "{group}"')
    for unit in df_group.unit.unique():
        df_unit = df_group.query(f'unit == {unit}')
        mask = (test_df.group == group) & (test_df.unit == unit)
        test_df.loc[mask, 'cycle_max'] = df_unit.cycle.max()
test_df.loc[:, 'RUL'] = test_df.cycle_max - test_df.cycle
```

```python
for group in test_df.group.unique():
    df_group = test_df.query(f'group == "{group}"')
    df_rul = pd.read_csv(
        os.path.join(data_path, group.replace('test', 'RUL')),
        index_col=False, names=['RUL']
    )
    for unit in df_group.unit.unique():
        mask = (test_df.group == group) & (test_df.unit == unit)
        test_df.loc[mask, 'RUL'] = test_df.loc[mask, 'RUL'] + df_rul.RUL[unit-1]
```

# Artificial Neural Network Modelling

```python
class TimeSeriesDataset(Dataset):
    def __init__(self, df, feature_columns, unit_column, group_column, target_column, sequence_length=20, step_size=1):
        self.df = df
        self.feature_columns = feature_columns
        self.unit_column = unit_column
        self.group_column = group_column
        self.target_column = target_column
        self.sequence_length = sequence_length
        self.step_size = step_size

        # Group by unit and group
        self.grouped_data = df.groupby([group_column, unit_column])

        # Store the start indices for each unit's sequences
        self.indices = self._compute_indices()

    def _compute_indices(self):
        """
        Computes start indices for each unit's time series, based on the step size and sequence length.
        """
        indices = []
        for (group, unit), unit_data in self.grouped_data:
            unit_data_length = len(unit_data)
            # Compute the valid start indices for each time series
            num_samples = (unit_data_length - self.sequence_length) // self.step_size + 1
            for i in range(num_samples):
                start_idx = i * self.step_size
                indices.append((group, unit, start_idx))
        return indices

    def __len__(self):
        return len(self.indices)

    def __getitem__(self, idx):
        """
        Retrieves the sample at the given index. Dynamically extracts the sequence from the data.
        """
        group, unit, start_idx = self.indices[idx]
        unit_data = self.grouped_data.get_group((group, unit))  # Retrieve the specific time series

        # Extract the features for the sequence
        features = unit_data[self.feature_columns].values[start_idx:start_idx + self.sequence_length]
        rul = unit_data[self.target_column].values[start_idx:start_idx + self.sequence_length].astype(int)
        return torch.tensor(features, dtype=torch.float32), torch.tensor(rul, dtype=torch.float32)
```
